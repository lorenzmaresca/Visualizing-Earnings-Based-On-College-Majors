## Part 2 - Questions:
1. What percent of majors are predominantly male? Predominantly female?
2. What's the most common median salary range?

#### Q1 - What percent of majors are predominantly male? Predominantly female?

```python
majors_count = recent_grads.groupby("Major_category", as_index = False)["Men", "Women"].sum()
majors_count
```

```python
majors_count["% Male"] = majors_count["Men"]/(majors_count["Men"]+ majors_count["Women"])
majors_count["% Female"] = majors_count["Women"]/(majors_count["Men"]+ majors_count["Women"])
majors_count["Boolean"] = majors_count["Men"] > majors_count["Women"]
sex = []
for row in majors_count["Boolean"]:
    if row == True:
        sex.append("Male")
    else:
        sex.append("Female")
majors_count["Predominant Sex"] = sex
```

```python
majors_count.drop(["Boolean"], inplace = True, axis = 1)
```

```python
majors_count
```

```python
table = majors_count[["Major_category", "Predominant Sex"]]
table
```
